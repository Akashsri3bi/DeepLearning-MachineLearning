```python
#Using Linear Regression 

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope , intercept , r , p , std_err = stats.linregress(x , y) #This gives us the lineof best fit for our gven x any y values 
#Create a function that uses the slope and intercept values to return a new value. This new value represents where on the y-axis the corresponding x value will be placed:
def regression(x):
  return slope*x + intercept

print(regression(10))
mymodel = list(map(regression , x))
plt.scatter(x,y)
plt.plot(x , mymodel) #PLOT DISPLays a line passing through x and y
plt.show()
```

```python
#Polynomial regression 

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

#making a polynomial model
myPolyModel = np.poly1d(np.polyfit(x,y,3)) #3 represents the degree 
print(myPolyModel)
myline = np.linspace(1,22,100)
plt.scatter(x,y) 
#Drawing the line of polynomial regression
plt.plot(myPolyModel , myPolyModel(myline))
plt.show() 
```
