```python
from matplotlib import pyplot
import numpy as np 
city = ('Pune' , 'Satara' , 'Mumbai' , 'Kanpur' , 'Bhopal' , 'Assam')
y_val = np.arange(len(city))
rank = [4,7,1,3,2,5]
explode = (0.2 , 0 , 0 , 0 , 0  ,0) # to set a portion of pie chart apart .

colors = ['yellowgreen' , 'pink'  , 'grey' , 'red' , 'orange']
pyplot.pie(rank , explode = explode , labels = city , colors = colors , autopct = '%1.1f%%', shadow=True , startangle=120)
pyplot.axis('equal')
pyplot.show()
```

```python
from matplotlib import pyplot
import numpy as np 

labels = ['G1' , 'G2' , 'G3' , 'G4' , 'G5']
men_means = [20,30,35,27,35]
women_means = [25,32,30,34,20] 
men_std = [2,3,4,1,2]
women_std = [3,5,2,3,3]
width = 0.35

fig,ax = pyplot.subplots()
ax.bar(labels , men_means , width , yerr = men_std , label = 'Men')
ax.bar(labels , women_means , width , yerr = women_std , bottom = men_means , label = 'Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

pyplot.show()
```
