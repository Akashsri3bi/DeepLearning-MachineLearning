```python
#@title Chatbot
import nltk
from nltk.chat.util import Chat,reflections
```

```python
reflections= {
  'i': 'you',
 'i am': 'you are',
 'i was': 'you were',
 "i'd": 'you would',
 "i'll": 'you will',
 "i'm": 'you are',
 "i've": 'you have',
 'me': 'you',
 'my': 'your',
 'you': 'me',
 'you are': 'I am',
 'you were': 'I was',
 "you'll": 'I will',
 "you've": 'I have',
 'your': 'my',
 'yours': 'mine'
 }
```

```python
pairs =[
       

        [
         
         r"my name is (.*)" , 
         ["Hello %1 , How are you Today ? "]

        ] , 
       
       [
        
        R"What is your name ? " ,
        ["I am a Bot created on your name , you can call me crazy"] 


       ], 

        [
         
         R"How are you ? ", 
         ["I am doing good how are you ?"]

        ] ,


       [
          R"quit", 
         ["Bye take care , see you real soon  /(-_-)/ "]
        
       ] , 
]
```

```python
def chat():
  print("Hello ,I  am chatbot created by your name for your service ")

  chat = Chat(pairs ,reflections)
  chat.converse()


if __name__=="__main__" :
  chat()
  
```
