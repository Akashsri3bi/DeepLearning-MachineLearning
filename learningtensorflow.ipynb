```python
#Core Learning algorithms 
from __future__ import absolute_import, division, print_function, unicode_literals

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc

import tensorflow as tf
```

```python
dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv') # training data
dfeval = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv') # testing data
print(dftrain.head())
y_train = dftrain.pop('survived') #This will contain the y axis data for our training data set
y_eval = dfeval.pop('survived') #This will contain the y axis data for our testing data set
print(dftrain.head())
print(y_train)
print()
#To find one specific entry in dataset we can use loc method 
print(dftrain.loc[0] , y_train.loc[0]) 
print()
#We can find the data for specific column also 
print(dftrain["age"])
print()

print(dfeval.shape)
```

```python
CATEGORICAL_COLUMNS = ['sex', 'n_siblings_spouses', 'parch', 'class', 'deck',
                       'embark_town', 'alone'] #A data which does not have numeric values instead is in form of text or String 
                       #So to do any mathematical work on this data we need to convert it into some mathematical form 


NUMERIC_COLUMNS = ['age', 'fare'] #This is the data which is in the form of numbers 
 
feature_columns = []
for feature_name in CATEGORICAL_COLUMNS:
  vocabulary = dftrain[feature_name].unique()  # gets a list of all unique values from given feature column
  #What this method is doing is finding the number of unique values in that particular column

  feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature_name, vocabulary))
  #This will make a column with our feature name which was actually sex or alone and connect it with different vocabulary values
  #The cryptic lines of code inside the append() create an object that our model can use to map string values like "male" and "female" to integers. This allows us to avoid manually having to encode our dataframes.

for feature_name in NUMERIC_COLUMNS:
  feature_columns.append(tf.feature_column.numeric_column(feature_name, dtype=tf.float32))

print(feature_columns)
```

```python
dftrain['sex'].unique()
```

```python
#Making an input function 

def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
  def input_function():  # inner function, this will be returned
    ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))  # create tf.data.Dataset object with data and its label
    if shuffle:
      ds = ds.shuffle(1000)  # randomize order of data
    ds = ds.batch(batch_size).repeat(num_epochs)  # split dataset into batches of 32 and repeat process for number of epochs
    return ds  # return a batch of the dataset
  return input_function  # return a function object for use

train_input_fn = make_input_fn(dftrain, y_train)  # here we will call the input_function that was returned to us to get a dataset object we can feed to the model
eval_input_fn = make_input_fn(dfeval, y_eval, num_epochs=1, shuffle=False)

linear_est = tf.estimator.LinearClassifier(feature_columns=feature_columns)
linear_est.train(train_input_fn)  # train
result = linear_est.evaluate(eval_input_fn)  # get model metrics/stats by testing on tetsing data

clear_output()  # clears consoke output
print(result['accuracy'])  # the result variable is simply a dict of stats about our model
print(result)
```

```python
result = list(linear_est.predict(eval_input_fn)) #We need to pass input function to predict
print(dfeval.loc[3])
#To actually check if that person survived or not and what our model gives the result which again depends upon the accuracy of the model
print(y_eval.loc[3])
print(result[3]['probabilities'][1]) # 1 represents that it survived  # and 0 in the list represents that it did'nt survive 
```

```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf 
import pandas as pd 
```

```python
CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']
# Lets define some constants to help us later on
```

```python
#Our dataset for this flowers model 
train_path = tf.keras.utils.get_file(
    "iris_training.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file(
    "iris_test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0) #Row zero is the header
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)
# Here we use keras (a module inside of TensorFlow) to grab our datasets and read them into a pandas dataframe
```

```python
train.head()
```

```python
train_y = train.pop('Species')
test_y = test.pop('Species')
train.head() # the species column is now gone
```

```python
train_y.head()
```

```python
#Input function 
def input_fn(features ,labels,training = True , batch_size = 256):
  #Converting the inputs into a dataset 
  dataset = tf.data.Dataset.from_tensor_slices(dict(features) , labels) 

  #Shuffling our dataset if we our in training mode 
  if training:
    dataset = dataset.shuffle(1000).repeat() 


  return dataset.batch(batch_size)

# Feature columns describe how to use the input.
my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
print(my_feature_columns)
```

```python
# Build a DNN with 2 hidden layers with 30 and 10 hidden nodes each.
classifier = tf.estimator.DNNClassifier( #This estimator module stores a bunch of different models/Classifiers from tensorflow
    feature_columns=my_feature_columns,
    # Two hidden layers of 30 and 10 nodes respectively.
    hidden_units=[30, 10],
    # The model must choose between 3 classes.
    n_classes=3)
```

```python
classifier.train(
    input_fn=lambda: input_fn(train, train_y, training=True))
# We include a lambda to avoid creating an inner function previously
```
